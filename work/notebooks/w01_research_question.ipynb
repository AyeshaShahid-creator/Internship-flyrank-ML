## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Question: Which pages should be reviewed first for refresh, expansion, or monitoring?
Unit of analysis: a page.
Decision: whether a page gets prioritized in the content team's review queue this cycle.
Who acts on it: a content editor or SEO team member, who would then refresh, expand, or re-optimize that page.
Action: pages ranked higher get worked on first; lower-ranked pages wait or get monitored.
Cost of a wrong call: flagging a page that didn't need it wastes limited review time on something that was already fine. Missing a page that was genuinely declining means it keeps losing visibility or clicks unnoticed, making recovery harder later. Since review time is limited, ranking by priority (not just yes/no) helps the team focus on the most urgent cases first.
